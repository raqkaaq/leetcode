### 23. Merge k Sorted Lists
Hard

You are given an array of k linked-lists lists, each linked-list is sorted in ascending order.

Merge all the linked-lists into one sorted linked-list and return it.

 

Example 1:

Input: lists = [[1,4,5],[1,3,4],[2,6]]
Output: [1,1,2,3,4,4,5,6]
Explanation: The linked-lists are:
[
  1->4->5,
  1->3->4,
  2->6
]
merging them into one sorted list:
1->1->2->3->4->4->5->6
Example 2:

Input: lists = []
Output: []
Example 3:

Input: lists = [[]]
Output: []
 

Constraints:

k == lists.length
0 <= k <= 104
0 <= lists[i].length <= 500
-104 <= lists[i][j] <= 104
lists[i] is sorted in ascending order.
The sum of lists[i].length will not exceed 104.

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
    def make(list):
        if len(list) == 0:
            return None
        node = ListNode(list[0], None)
        head = node
        for i in range(1,len(list)):
            node.next = ListNode(list[i], None)
            node = node.next
        return head
    def __str__(self):
        node = self
        stri = '['
        while node != None:
            stri += str(node.val)
            if node.next != None:
                stri += ','
            node = node.next
        stri += ']'
        return stri
from typing import Optional
#This works, the issue is that it checks each lists[i] for every time it puts a new one
class Solution:
    def mergeKLists(self, lists: list[Optional[ListNode]]) -> Optional[ListNode]:
        k = len(lists)
        if k == 0 or lists.count(None) == k:
            return None
        dummy = ListNode(0, None)
        p = dummy
        while lists.count(None) < k:
            mini = float('inf')
            index = 0
            for i in range(k):
                if not lists[i]:
                    continue
                elif lists[i].val < mini:
                    mini = lists[i].val
                    index = i
            lists[index] = lists[index].next
            p.next = ListNode(mini, None)
            p = p.next
        return dummy.next
```

```python
sol = Solution()
l = [[1,4,5],[1,3,4],[2,6]]
for i,val in enumerate(l):
    l[i] = ListNode.make(val)
print(sol.mergeKLists(l))
l = [[]]
for i,val in enumerate(l):
    l[i] = ListNode.make(val)
print(sol.mergeKLists(l))
l = []
for i,val in enumerate(l):
    l[i] = ListNode.make(val)
print(sol.mergeKLists(l))
l = [[1,3,4], [12,13,15],[2,16]]
for i,val in enumerate(l):
    l[i] = ListNode.make(val)
print(sol.mergeKLists(l))
```

```python
#this has bad space complexity since it has to create a list before converting it to a linked list
#possibly could create a list nodes that is a snapshot of the current heads and take the minimum
class Solution:
    def mergeKLists(self, lists: list[Optional[ListNode]]) -> Optional[ListNode]:
        k = len(lists)
        if k == 0 or lists.count(None) == k:
            return None
        dummy = ListNode(0, None)
        p = dummy
        nodes = []
        for l in lists:
            while l:
                nodes.append(l.val)
                l = l.next
        nodes.sort()
        for n in nodes:
            p.next = ListNode(n, None)
            p = p.next
        return dummy.next
```
