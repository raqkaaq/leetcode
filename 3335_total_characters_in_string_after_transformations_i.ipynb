# 3335. Total Characters in String After Transformations I
Medium

Hint
You are given a string s and an integer t, representing the number of transformations to perform. In one transformation, every character in s is replaced according to the following rules:

If the character is 'z', replace it with the string "ab".
Otherwise, replace it with the next character in the alphabet. For example, 'a' is replaced with 'b', 'b' is replaced with 'c', and so on.
Return the length of the resulting string after exactly t transformations.

Since the answer may be very large, return it modulo 109 + 7.

 

Example 1:

Input: s = "abcyy", t = 2

Output: 7

Explanation:

First Transformation (t = 1):
'a' becomes 'b'
'b' becomes 'c'
'c' becomes 'd'
'y' becomes 'z'
'y' becomes 'z'
String after the first transformation: "bcdzz"
Second Transformation (t = 2):
'b' becomes 'c'
'c' becomes 'd'
'd' becomes 'e'
'z' becomes "ab"
'z' becomes "ab"
String after the second transformation: "cdeabab"
Final Length of the string: The string is "cdeabab", which has 7 characters.
Example 2:

Input: s = "azbk", t = 1

Output: 5

Explanation:

First Transformation (t = 1):
'a' becomes 'b'
'z' becomes "ab"
'b' becomes 'c'
'k' becomes 'l'
String after the first transformation: "babcl"
Final Length of the string: The string is "babcl", which has 5 characters.
 

Constraints:

1 <= s.length <= 105
s consists only of lowercase English letters.
1 <= t <= 105

```python
import numpy as np
class Solution:
    def lengthAfterTransformations(self, s: str, t: int) -> int:
        freq = np.zeros(26, dtype=int)
        for c in s:
            freq[ord(c) - ord('a')] += 1
        t_mat = np.zeros((26, 26), dtype=object)
        for i in range(25):
            t_mat[i][i+1] = 1
        t_mat[25][0], t_mat[25][1] = 1, 1
        t_mat = np.linalg.matrix_power(t_mat, t)
        freq = freq @ t_mat
        return sum(freq) % (10**9 + 7)
```

```python
#another solution
class Solution:
    def lengthAfterTransformations(self, s: str, t: int) -> int:
        freq, overflow = [0]*26, 10**9 + 7
        for c in s:
            freq[ord(c) - ord('a')] += 1
        for _ in range(t):
            tp = freq[25]
            for i in range(25, 0, -1):
                freq[i] = freq[i-1]
            freq[0] = tp
            freq[1] += tp
        return sum(freq) % overflow
#10**9 + 7 is chosen since its a very large prime number, so supposing that the answer is less than that, we can use modulo to keep the number small and avoid overflow
```

```python
sol = Solution()
s, t = 'abcyy', 2
assert sol.lengthAfterTransformations(s, t) == 7
s, t = "azbk", 1
assert sol.lengthAfterTransformations(s, t) == 5
s, t = "zz", 7
assert sol.lengthAfterTransformations(s, t) == 4
s,t = "jqktcurgdvlibczdsvnsg", 7517
assert sol.lengthAfterTransformations(s, t) == 79033769
```
